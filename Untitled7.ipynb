```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 3, 7]
y = [2, 6, 14]

plt.scatter(x, y, color='blue')

w = 0
b = 0
lr = 0.01

for i in range(1000):
  y_pred = np.array(x) * w + b
error = np.mean((np.array(y) - y_pred)**2)
w_1 = w + lr * 2 * np.mean((y - y_pred) * np.array(x))
b_1 = b + lr * 2 * np.mean(y - y_pred)
w_2 = w_1 + lr * 2 * np.mean((y - y_pred) * np.array(x))
b_2 = b_1 + lr * 2 * np.mean(y - y_pred)
w_3 = w_2 + lr * 2 * np.mean((y - y_pred) * np.array(x))
b_3 = b_2 + lr * 2 * np.mean(y - y_pred)

x_values = [0, 7]
y_values = [w_3*x + b_3 for x in x_values]
plt.plot(x_values, y_values)

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
